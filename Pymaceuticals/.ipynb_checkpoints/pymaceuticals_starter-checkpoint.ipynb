## Observations and Insights

```python
# Mean tumor volumes across all data points were markedly smaller for those subjects given regimens of Capomulin and Ramicane
# when compared to those who received a placebo. 
#
# Amongst the mice who received Capomulin, we can observe a Pearson's correlation coefficient of 0.84 for the bivariate data 
# of average tumor volume (mm3) and weight (g) for the studied mice. The positive correlation indicates that the heavier mice 
# may be more likely to develop larger tumors.
#
# As can be observed in the box plot of the final tumor volume of each mouse comparing Capomulin, Ramicane, Infubinol, 
# and Ceftamin, there is an overlap in interquartile range for Capomulin and Ramicane and for Infubinol and Ceftamin.
# The volumes measured for the former drugs is substantially smaller than for the latter.
```

## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset

capomulin_merge_df = pd.merge(study_results, mouse_metadata, on='Mouse ID', how='inner')
capomulin_merge_df.head(15)
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM 
# of the tumor volume for each regimen

regimen_sub = capomulin_merge_df.loc[:,['Drug Regimen','Tumor Volume (mm3)']]
regimen_stats_df = pd.DataFrame({
    'Tumor Volume (mm3) Mean' : regimen_sub.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean(),
    'Tumor Volume (mm3) Median' : regimen_sub.groupby('Drug Regimen')['Tumor Volume (mm3)'].median(),
    'Tumor Volume (mm3) Variance' : regimen_sub.groupby('Drug Regimen')['Tumor Volume (mm3)'].var(ddof = 0),
    'Tumor Volume (mm3) StdDev' : regimen_sub.groupby('Drug Regimen')['Tumor Volume (mm3)'].std(ddof = 0),
    'Tumor Volume (mm3) SEM' : regimen_sub.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
})
regimen_stats_df
```

```python
# Utilize the aggregate function <agg()> to generate a summary statistics table of 
# mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# as above. (reduces the amount of dataframe coding necessary)

regimen_stats_df_agg_test = capomulin_merge_df.loc[:,['Drug Regimen','Tumor Volume (mm3)']]
regimen_stats_df_agg_test.groupby('Drug Regimen').agg(['mean', 'median', 'var', 'std', 'sem'])
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens.
# Calculate the IQR and quantitatively determine if there are any potential outliers.

# Sort the data on Timepoint descending to place the greatest data point for each mouse in the highest position
# then drop_duplicates on Mouse ID to keep only the last Timepoint value for each mouse.
maxmouse = capomulin_merge_df.sort_values('Timepoint', ascending=False).drop_duplicates(['Mouse ID'])

# Extract data for the four of the most promising treatment regimens and sort them by Drug Regimen and Tumor Volume in ascending order
maxtime = maxmouse[(maxmouse['Drug Regimen']=='Capomulin') | 
                  (maxmouse['Drug Regimen']=='Ramicane') |
                  (maxmouse['Drug Regimen']=='Infubinol') |
                  (maxmouse['Drug Regimen']=='Ceftamin')].sort_values(by=['Drug Regimen', 'Tumor Volume (mm3)'], ascending=True)

# Extract a list for each of the Drug Regimens because the arrays are not of the same length.
capo_tumors = maxtime[(maxtime['Drug Regimen']=='Capomulin')]['Tumor Volume (mm3)'].tolist()
rami_tumors = maxtime[(maxtime['Drug Regimen']=='Ramicane')]['Tumor Volume (mm3)'].tolist()
infu_tumors = maxtime[(maxtime['Drug Regimen']=='Infubinol')]['Tumor Volume (mm3)'].tolist()
ceft_tumors = maxtime[(maxtime['Drug Regimen']=='Ceftamin')]['Tumor Volume (mm3)'].tolist()

capo_lowerq = np.quantile(capo_tumors, .25)
capo_q2 = np.quantile(capo_tumors, .5)
capo_upperq = np.quantile(capo_tumors, .75)
capo_iqr = capo_upperq-capo_lowerq
capo_lower_bound = capo_lowerq - (1.5*capo_iqr)
capo_upper_bound = capo_upperq + (1.5*capo_iqr)

rami_lowerq = np.quantile(rami_tumors, .25)
rami_q2 = np.quantile(rami_tumors, .5)
rami_upperq = np.quantile(rami_tumors, .75)
rami_iqr = rami_upperq-rami_lowerq
rami_lower_bound = rami_lowerq - (1.5*rami_iqr)
rami_upper_bound = rami_upperq + (1.5*rami_iqr)

infu_lowerq = np.quantile(infu_tumors, .25)
infu_q2 = np.quantile(infu_tumors, .5)
infu_upperq = np.quantile(infu_tumors, .75)
infu_iqr = infu_upperq-infu_lowerq
infu_lower_bound = infu_lowerq - (1.5*infu_iqr)
infu_upper_bound = infu_upperq + (1.5*infu_iqr)

ceft_lowerq = np.quantile(ceft_tumors, .25)
ceft_q2 = np.quantile(ceft_tumors, .5)
ceft_upperq = np.quantile(ceft_tumors, .75)
ceft_iqr = ceft_upperq-ceft_lowerq
ceft_lower_bound = ceft_lowerq - (1.5*ceft_iqr)
ceft_upper_bound = ceft_upperq + (1.5*ceft_iqr)

print(f"Capomulin's median final tumor volume was {round(capo_q2,3)}. Values below {round(capo_lower_bound,3)} or above {round(capo_upper_bound,3)} could be outliers.")
print(f"Ramicane's median final tumor volume was {round(rami_q2,3)}. Values below {round(rami_lower_bound,3)} or above {round(rami_upper_bound,3)} could be outliers.")
print(f"Infubinol's median final tumor volume was {round(infu_q2,3)}. Values below {round(infu_lower_bound,3)} or above {round(infu_upper_bound,3)} could be outliers.")
print(f"Ceftamin's median final tumor volume was {round(ceft_q2,3)}. Values below {round(ceft_lower_bound,3)} or above {round(ceft_upper_bound,3)} could be outliers.")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
# Append together the lists for each of the Drug Regimens created previously because the arrays are not of the same length.
tumors = [capo_tumors, rami_tumors, infu_tumors, ceft_tumors]

markerstyle = dict(markerfacecolor='xkcd:emerald', marker='p', markersize=10)

fig1, ax1 = plt.subplots(figsize=(16,10))
ax1.set_title('Final Tumor Volume for Most Promising Regimens', 
              fontdict={'fontsize': 24,
                        'fontweight' : 20,
                        'verticalalignment': 'baseline',
                        'horizontalalignment': 'center'})
ax1.boxplot(tumors, notch=True, flierprops=markerstyle)
plt.xticks([1, 2, 3, 4], ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])
plt.xlabel('Drug Regimen', labelpad=16, fontsize=16)
plt.ylabel('Tumor Volume (mm3)', labelpad=16, fontsize=16)
plt.show()
```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin

capo_mice_df = capomulin_merge_df[(capomulin_merge_df['Drug Regimen']=='Capomulin')]
capo_single_df = capo_mice_df[(capo_mice_df['Mouse ID']=='r554')]

# Set x axis and tick locations
x_axis = np.arange(len(capo_single_df))
tick_locations = [value for value in x_axis]

# Create a list indicating where to write x labels and set figure size to adjust for space
fig5, ax5 = plt.subplots(figsize=(16,10))
ax5.set_title('Final Tumor Volume for Most Promising Regimens', fontsize=24)
plt.plot(x_axis, capo_single_df['Tumor Volume (mm3)'], color='xkcd:orangered', alpha=0.8, lw=5)
plt.xticks(tick_locations, capo_single_df['Timepoint'], rotation='horizontal')
plt.xlabel('Timepoint of Measurement', labelpad=16, fontsize=16)
plt.ylabel('Tumor Volume (mm3)', labelpad=16, fontsize=16)
plt.grid(b=True, which='major', axis='both', alpha=0.3)
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
capo_avg_vol = pd.DataFrame({
    'Average Tumor Volume (mm3)' : capo_mice_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean(),
    'Weight (g)' : capo_mice_df.groupby('Mouse ID')['Weight (g)'].mean()
})
# Create the scatter plot from the data frame.
capo_avg_vol.plot(figsize=(16,10), kind='scatter', x='Weight (g)', y='Average Tumor Volume (mm3)', c='DarkBlue', s=28, grid='major')
plt.title('Capomulin: Avg Tumor Volume vs Mouse Weight', fontsize=24, y=1.02)
plt.xlabel('Weight (g)', labelpad=16)
plt.ylabel('Average Tumor Volume (mm3)', labelpad=16)
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume 
# for the Capomulin regimen

weight = capo_avg_vol.loc[:,'Weight (g)']
avgtumor = capo_avg_vol.loc[:,'Average Tumor Volume (mm3)']

# Compute the Pearson correlation coefficient between Weight and Average Tumor Volume
correlation = st.pearsonr(weight,avgtumor)

# Perform a linear regression on Weight versus Average Tumor Volume
capo_slope, capo_int, capo_r, capo_p, capo_std_err = st.linregress(weight, avgtumor)

# Create equation of line to calculate predicted Average Tumor Volume
capo_fit = capo_slope * weight + capo_int

fig6, ax6 = plt.subplots(figsize=(16,10))
ax6.set_title('Capomulin: Avg Tumor Volume vs Mouse Weight', fontsize=24)

plt.scatter(weight,avgtumor, c='DarkBlue', s=28)
plt.plot(weight,capo_fit,'r:')
plt.xlabel('Weight (g)', labelpad=16)
plt.ylabel('Average Tumor Volume (mm3)', labelpad=16)
plt.grid(b=True, which='major', axis='both', alpha=0.3)

plt.show()

print(f"The correlation between both factors is {round(correlation[0],2)}")
```
